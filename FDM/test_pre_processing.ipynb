```python
from BaselineRemoval import BaselineRemoval
import numpy as np
```

```python
wavelength_list = []
count_list = []
with open ('3rep_a.txt', 'r')as origin_file:
    with open('raw_data.txt', 'a') as r:
        for l in origin_file:
            count_list.append(float(l.rsplit('	', 1)[1]))
            wavelength_list.append(float(l.rsplit('	', 1)[0]))
            
            
```

```python
baseObj=BaselineRemoval(count_list)
Zhangfit_output=baseObj.ZhangFit()
m_o_data=baseObj.ModPoly(2)
Imodpoly_output=baseObj.IModPoly(2)
```

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt


plt.figure(figsize = (40,20))
plt.plot(wavelength_list, Zhangfit_output, linewidth = 0.5)
plt.xlabel('wavelength')
plt.ylabel('counts')
plt.xticks(range( int(wavelength_list[0]), int(wavelength_list[-1]), 99 ))
plt.show()
```
